```python
import numpy as np
from sklearn.model_selection import LeaveOneGroupOut

y =dataEmdat["answer"]
X =dataEmdat.drop(["answer"], axis= 1)

groups = dataEmdat["Part_id"]
logo = LeaveOneGroupOut()
logo.get_n_splits(X, y, groups)

logo.get_n_splits(groups=groups)
print(logo)
LeaveOneGroupOut()
# for train_index, test_index in logo.split(X, y, groups):
#     print("TRAIN:", train_index, "TEST:", test_index)
#     X_train, X_test = X[train_index], X[test_index]
#     y_train, y_test = y[train_index], y[test_index]
#     print(X_train, X_test, y_train, y_test)
for train_index, test_index in logo.split(X, y, groups):
    print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    print(X_train, X_test, y_train, y_test)
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_validate
from statistics import mean
from sklearn.ensemble import RandomForestClassifier
random_forest = RandomForestClassifier()


scores = ["accuracy", "precision", "recall", "f1"]
scores = cross_validate(random_forest, X, y, cv=logo.split(X, y, groups), n_jobs=2, scoring = scores, error_score="raise")
print("Accuracy = {}".format(mean(scores["test_accuracy"])))
print("Precision = {}".format(mean(scores["test_precision"])))
print("Recall = {}".format(mean(scores["test_recall"])))
print("F1 = {}".format(mean(scores["test_f1"])))
```

```python
from sklearn import metrics
from sklearn.model_selection import cross_validate
from statistics import mean
from sklearn.dummy import DummyClassifier
dummy_clf = DummyClassifier(strategy="most_frequent")

scores = ["accuracy", "precision", "recall", "f1"]
scores = cross_validate(dummy_clf, X, y, cv=logo.split(X, y, groups), n_jobs=2, scoring = scores, error_score="raise")
print("Accuracy = {}".format(mean(scores["test_accuracy"])))
print("Precision = {}".format(mean(scores["test_precision"])))
print("Recall = {}".format(mean(scores["test_recall"])))
print("F1 = {}".format(mean(scores["test_f1"])))
```

```python

l=0
for i in range(16):
    k=0
    for j in range(3):
        seg_total=dfo[i][(dfo[i]["Recording timestamp"].values >tm[i][k] )&(dfo[i]["Recording timestamp"].values <tm[i][k+1])]
        seg_screen=len(seg_total[seg_total["gaze_direction"]=="screen"])
        seg_notes=len(seg_total[seg_total["gaze_direction"]=="down"])
        seg_none=len(seg_total[seg_total["gaze_direction"]=="None"])
        if len(seg_total) != 0 :
            dataEmdat2.loc[(dataEmdat2["Part_id"]==i+1) & (dataEmdat2["notes"].index==j+l),"notes"]=int((seg_notes*100)/len(seg_total))
            dataEmdat2.loc[(dataEmdat2["Part_id"]==i+1) & (dataEmdat2["screen"].index==j+l),"screen"]=int((seg_screen*100)/len(seg_total))
        else:
            dataEmdat2.loc[(dataEmdat2["Part_id"]==i+1) & (dataEmdat2["notes"].index==j+l),"notes"]=0
            dataEmdat2.loc[(dataEmdat2["Part_id"]==i+1) & (dataEmdat2["screen"].index==j+l),"screen"]=0
        k=k+2
    l=l+len(dataEmdat2.loc[(dataEmdat2["Part_id"]==i+1)])
dataEmdat2.head(12)
```
