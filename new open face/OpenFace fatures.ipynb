```python
# same code but improved and added sum
import pandas as pd
import numpy as np

# read the tsv file into a pandas DataFrame
output_df = pd.read_csv("openface30.tsv", sep="\t")

# loop over each participant's csv file
for i in range(1, 26):
    Part_id = "p" + str(i)

    # read the csv file for the current participant into a pandas DataFrame
    participant_df = pd.read_csv(f"C:/Users/Sina/Desktop/Stage V2/Sripts/new open face/open face data/{participant}.csv")
    
    # Filter data where success column value is 1
    participant_df = participant_df[participant_df[' success'] == 1]

    # convert the "timestamp" column to microseconds
    participant_df[" timestamp"] *= 1000000

    # find the rows in the tsv file that correspond to the current participant
    rows = output_df["Part_id"] == i

    # loop over the relevant rows in the tsv file
    for index, row in output_df[rows].iterrows():
        start_time = row["Start Time"]
        end_time = row["End Time"]

        # find the rows in the current participant's csv file that correspond to the current time range
        time_rows = (participant_df[" timestamp"] >= start_time) & (participant_df[" timestamp"] <= end_time)

        # calculate the mean, standard deviation, minimum, maximum, and sum
        columns = [" gaze_angle_x", " gaze_angle_y", " pose_Tx", " pose_Ty", " pose_Tz",
                   " AU01_r", " AU02_r", " AU04_r", " AU05_r", " AU06_r", " AU07_r",
                   " AU09_r", " AU10_r", " AU12_r", " AU14_r", " AU15_r", " AU17_r",
                   " AU20_r", " AU23_r", " AU25_r", " AU26_r", " AU45_r"]
        
        for column in columns:
            values = participant_df[time_rows][column]
            
            mean_value = np.mean(values)
            std_value = np.std(values)
            min_value = np.min(values)
            max_value = np.max(values)
            sum_value = np.sum(values)
            
            output_df.at[index, f"Mean{column}"] = mean_value
            output_df.at[index, f"Standard Deviation{column}"] = std_value
            output_df.at[index, f"Minimum{column}"] = min_value
            output_df.at[index, f"Maximum{column}"] = max_value
            output_df.at[index, f"Sum{column}"] = sum_value
            
# write the updated tsv file back to disk
output_df.to_csv("openface30.tsv", sep="\t", index=False)
```
